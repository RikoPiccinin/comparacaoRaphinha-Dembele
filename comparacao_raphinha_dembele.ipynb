```python
import pandas as pd
```

```python
df = pd.read_excel('/content/comparacao.xlsx')
```

```python
df
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(6,6))

bars = plt.bar(df['Estatisticas'], df['Valores'], color='skyblue')

plt.title("Comparação Raphinha x Dembele\n Numeros Raphinha", color='blue')
plt.xlabel('Estatísticas', color='green')
plt.ylabel('Valores', color='gray')
plt.xticks(rotation=20)

# Adicionar valores acima das barras
for bar, valor in zip(bars, df["Valores"]):
    if float(valor).is_integer():
        texto = f"{int(valor)}"      # remove ".0"
    else:
        texto = f"{valor:.2f}"       # mostra 2 casas decimais
    plt.text(
        bar.get_x() + bar.get_width()/2,
        bar.get_height() + 0.3,
        texto,
        ha="center", va="bottom", fontsize=10
    )

plt.tight_layout()

plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import pandas as pd



estatisticas = df["Estatisticas"]
x = range(len(estatisticas))
largura = 0.35

fig, ax = plt.subplots(figsize=(12,6))

bars1 = ax.bar([i - largura/2 for i in x], df["Valores"], width=largura, label="Raphinha", color="blue")
bars2 = ax.bar([i + largura/2 for i in x], df2["Valores"], width=largura, label="Dembele", color="red")

ax.set_xticks(x)
ax.set_xticklabels(estatisticas, rotation=30, ha="right")
ax.set_title("Comparação Raphinha x Dembele", color='#4e3a3b')
ax.set_ylabel("Valores")
ax.legend()

# Adicionar valores no topo das barras
def format_val(valor):
    if float(valor).is_integer():
        return str(int(valor))
    else:
        return f"{valor:.2f}"

for bar in bars1 + bars2:  # percorre todas as barras
    altura = bar.get_height()
    ax.text(
        bar.get_x() + bar.get_width()/2,  # posição x central
        altura + 0.3,                    # um pouco acima da barra
        format_val(altura),              # formata valor
        ha="center", va="bottom", fontsize=9, color="black"
    )


# 🔹 Adicionar logo
logo = mpimg.imread("/content/logo.png")  # substitua pelo caminho do seu logo
# posição: (x=0.8, y=0.8) -> canto superior direito
newax = fig.add_axes([0.45, 0.70, 0.15, 0.15], anchor='NE', zorder=1)
newax.imshow(logo)
newax.axis("off")

plt.tight_layout()
plt.show()
```
